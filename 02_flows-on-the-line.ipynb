# 2.2 Fixed point and stability

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

## 2.2.1

```python
f = lambda x, _: 4*x*x - 16
t = np.linspace(0, 0.5)
xs0 = np.linspace(-3, 2.2)
xs = [odeint(f, x0, t) for x0 in xs0] 
```

```python
for x in xs:
    plt.plot(t, x, 'k')
plt.ylim(-3.5,3.5)
```

## 2.2.7

```python
f = lambda x, _: np.exp(x) - np.cos(x)
t = np.linspace(0, 10)
xs0 = np.linspace(-8, 0.1)
xs = [odeint(f, x0, t) for x0 in xs0] 

for x in xs:
    plt.plot(t, x, 'k')
plt.ylim(-8,0.5)
```
